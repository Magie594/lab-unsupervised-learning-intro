```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000) 
    return tracks
```

```python
all_tracks = get_playlist_tracks("4rnleEAOdmFAbRcNCgZMpY")
len(all_tracks)
```

```python
import pandas as pd
from pandas import json_normalize

tracks2= json_normalize(all_tracks)
```

```python
tracks2
```

```python
def list_to_dict(x):
    return {i: x[i] for i in range(len(x))}
```

```python
tracks2['artist_dict'] = tracks2['track.artists'].apply(list_to_dict)
tracks2.head()
```

```python
tracks2['artist_dict']
```

```python
def expand_list_dict(row):
    df = pd.DataFrame.from_dict(row['artist_dict'], orient='index')
    df['song_id'] = row['track.id']
    return df

tracks2['artists_dfs'] = tracks2.apply(expand_list_dict, axis=1)
tracks2['artists_dfs'][0]
```

```python
artist_df = pd.DataFrame(columns=['external_urls', 'href', 'id', 'name', 'type', 'uri', 'song_id'])
for mini_df in tracks2['artists_dfs']:
    artist_df = pd.concat([artist_df, mini_df], axis=0)
```

```python
artist_df
```

```python
df_merged = pd.merge(left=tracks2,
                    right=artist_df,
                    how='inner',
                    left_on='track.id',
                    right_on='song_id')
```

```python
df =df_merged[['track.name', 'song_id', 'name', 'id']]
```
